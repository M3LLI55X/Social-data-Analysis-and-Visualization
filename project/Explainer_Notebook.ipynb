## Motivation
* **What is your dataset?**

  The data set that we use is [Motor Vehicle Collisions - Crashes](https://data.cityofnewyork.us/Public-Safety/Motor-Vehicle-Collisions-Crashes/h9gi-nx95), which is hosted by the City of New York. The city has an open data platform found [here](https://opendata.cityofnewyork.us/) and they update their information according the amount of data that is brought in. The NYC motor vehicle collision data set, contains up-to-date collision record ever since July 1st, 2012. This data set contains 1.98 million rows and 29 columns, and each record shows the date, time, location, the number of injured and/or killed people (both totally and in terms of pedestrian, cyclist, and motorist, respectively), along with the causes and involved types of vehicles, etc.

* **Why did you choose this/these particular dataset(s)?**

  Road safety is a critical issue, and is relevant to everybody's daily life. Therefore, it is important to look at the past collision history data and see what we can learn from the data to help avoid collisions in the future.

* **What was your goal for the end user's experience?**

  Combined with the visual results we provide, users can easily understand the facts and analysis we present on this topic in the text. In addition, the user should be able to clearly know the content of each visual analysis, and intuitively see the internal logical relationship.
