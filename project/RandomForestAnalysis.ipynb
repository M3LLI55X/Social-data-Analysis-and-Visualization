```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import os
import warnings; warnings.filterwarnings(action='once')
```

```python
df = pd.read_csv("C:\\Users\\hjjyc\\Desktop\\Motor.csv", dtype=str)
```

```python
# Missing values in columns
df.isna().sum()
```

```python
# Replace NAN for latitude and longitude with U
df['LATITUDE'].fillna('U',inplace=True) 
df['LONGITUDE'].fillna('U',inplace=True) 
```

```python
df.head()
```

```python
train=df[(df['LATITUDE']!='U')&(df['LONGITUDE']!='U')&(df['BOROUGH'].notnull())]
test=df[(df['LATITUDE']!='U')&(df['LONGITUDE']!='U')&(df['BOROUGH'].isnull())]
U=df[(df['LATITUDE']=='U')|(df['LONGITUDE']=='U')]
```

```python
from datetime import date
from sklearn.ensemble import RandomForestClassifier

cols = ['LATITUDE', 'LONGITUDE']
 
reg_ngk = RandomForestClassifier(random_state=100)
reg_ngk.fit(train[cols], train['BOROUGH'])

test['BOROUGH_RF'] = reg_ngk.predict(test[cols])
```

```python
import matplotlib.pyplot as plt

# Define the x and y variables
x = test['LONGITUDE']
y = test['LATITUDE']

# Define the color for each borough
colors = {'QUEENS': 'red', 'BROOKLYN': 'blue', 'MANHATTAN': 'green', 'BRONX': 'purple', 'STATEN ISLAND': 'orange'}

# Define the color for each data point based on its predicted borough
color_list = [colors[borough] for borough in test['BOROUGH_RF']]

plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')

# Create the scatter plot
# fig, ax = plt.subplots(figsize=(20, 20))
plt.scatter(x, y, color=color_list, alpha=0.5)

# Set the plot title and axis labels
# plt.set_title('Predicted Boroughs', fontsize=20)
# plt.set_xlabel('Longitude', fontsize=16)
# plt.set_ylabel('Latitude', fontsize=16)


# Decorations
plt.gca().set(xlim=(-74.28, -73.65), ylim=(40.48, 40.93),
              xlabel='LONGITUDE', ylabel='LATITUDE')

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("BOROUGH vs LONGITUDE/LATITUDE", fontsize=22)
plt.legend(fontsize=12)    
plt.show() 
```
